Machine Learning
Steps to creating the Machine Leaning Model.

Step 1:
Gathering the data, uploading the data

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

Step 2: data Cleaning/data wrangling

```python
#converting categorical varibles in to numerical format, by using the - one Hot Encoding.
data['sex']=data['sex'].apply({'female':1,'male':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':2,'northeast':3,'northwest':4}.get)
```

```python
data.head()
```

Step 3: Divinding the data into dependent variable, and indipendent variables.

```python
x=data[['age','sex','bmi','children','smoker','region']]
y=data['expenses']
```

Step 4: Splitting the data into Training and Testing

```python
from sklearn.model_selection import train_test_split
x_tarin,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8) #80% of data used for training(learning)
```

Creating our SMLM Model using Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression() #we activating our algorithm
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
#accuracy of SMLM
regression.score(x,y)
```
